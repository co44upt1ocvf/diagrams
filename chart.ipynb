```python
anime = pd.read_csv('anime.csv')
plt.bar(anime['Name'].head(10), anime['Episodes'].head(10), label = 'Рэйтинг', color = 'purple')
plt.legend()
plt.show()
```

Это столбчатая диаграмма соотношения эпизодов, касательно названия. Сначала читается файл, затем данные выбираются по первым 10 названиям в файле. Снизу подписаны названия тайтлов, а слева - эпизоды. plt.legend() показывает легенду с описаниями элементов графика, а plt.show() саму диаграмму.

```python
anime = pd.read_csv('anime.csv')
data = anime['Episodes'].value_counts().head(15)
plt.pie(data,labels=data.index,autopct='%1.1f%%', colors = ['pink','red'])
plt.legend()
plt.show()
```

Это круговая диаграмма, которая содержит в себе процентное соотношение кол-ва эпизодов. Сначала читается файл, затем данные выбираются по первым 15 названиям в файле, заполняется список информации с помощью value_counts(). Autopct='%1.1f%%' значит, что данные будут представлены в процентах.

```python
anime = pd.read_csv('anime.csv')
data = anime['Type'].value_counts().head(100)
plt.pie(data,labels=data.index,autopct='%1.1f%%', colors = ['purple','blue'])
plt.legend()
plt.show()
```

Это круговая диаграмма, которая содержит в себе процентное соотношение типов. Сначала читается файл, затем данные выбираются по первым 15 названиям в файле, заполняется список информации с помощью value_counts(). Autopct='%1.1f%%' значит, что данные будут представлены в процентах.
